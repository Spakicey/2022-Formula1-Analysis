```python
# Import dependencies
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import re
import time, datetime

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# import uncleaned .csv file
master_df_copy = pd.read_csv("master_1.csv")
master_df_copy.head(5)
```

```python
tod = master_df_copy['Time of day']
tod = tod.copy()
for count, race in enumerate(tod):
    # Race time is always first index
    if (count == 0):
        print(f"COUNT: {count} | VALUE: {race} | IN COUNT 0")
        end_time = race
        my_list = end_time.split(':')
        time_f = timedelta(hours=int(my_list[0]),minutes=int(my_list[1]), seconds=float(my_list[2])).total_seconds()
        tod.iloc[count] = time_f
    elif (count>0):
        print(f"COUNT: {count} | VALUE: {race} | ELIF")
        tod.iloc[count] = timedelta(hours=int(my_list[0]),minutes=int(my_list[1]), seconds=float(my_list[2])).total_seconds()
```

```python
race_time_copy = master_df_copy['Total Pit Time']
#race_time_copy = race_time.copy()
for count, race in enumerate(race_time_copy):
    # Race time is always first index
    if (count == 0):
        print(f"COUNT: {count} | VALUE: {race} | IN COUNT 0")
        end_time = race
        my_list = end_time.split(':')
        time_f = timedelta(minutes=int(my_list[1]), seconds=float(my_list[2])).total_seconds()
        race_time_copy.iloc[count] = time_f
    # If race contains substring 'lap' indicating 1+ Lap. Will add the number of laps indicated so 1+ Lap 
    # Will add 1 whole lap, 2+ Lap will add two laps, etc.
    elif 'lap' in race:
        print(f"COUNT: {count} | VALUE: {race} | IN LAP")
        race_time_copy.iloc[count] = time_f+(int(race[0])*time_f)
    # Use regex to check if a float value exists
    elif re.match(r'^-?\d+(?:\.\d+)$', race) is not None:
        print(f"COUNT: {count} | VALUE: {race} | IN INSTANCE")
        race_time_copy.iloc[count] = (time_f + float(race))
    # Anything else such as DNF, DSQ will get 999999
    else:
        print(f"COUNT: {count} | VALUE: {race} | IN ELSE")
        race_time_copy.iloc[count] = 999999
```
