```python
# Import dependencies
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.firefox import GeckoDriverManager
```

```python
# Set up splinter
executable_path = {'executable_path': GeckoDriverManager().install()}
browser = Browser('firefox', **executable_path, headless=False)
```

```python
# Visit Formula 1 site
url = 'https://www.formula1.com/en/results.html'
browser.visit(url)
# Optional delay for loading the page
browser.is_element_not_present_by_css('div.resultsarchive-filter-wrap', wait_time=1)
```

```python
# Set up html parser
html = browser.html
news_soup = soup(html, 'html.parser')
slide_elem = news_soup.select_one('div.resultsarchive-filter-container')
```

```python
# Click accept cookies
browser.find_by_id('truste-consent-button').click()
```

```python
bahrain_race_result_partial_link = slide_elem.findAll('a', class_='resultsarchive-filter-item-link FilterTrigger')[-22].get('href')
bahrain_race_result_partial_link
```

```python
# Click bahrain link
browser.links.find_by_partial_href(bahrain_race_result_partial_link).click()
```

```python
# Get race results link
bahrain_race_result_url = f'https://www.formula1.com{bahrain_race_result_partial_link}'
bahrain_race_result_url
```

```python
# Grab first table of webpage and create DF from it
race_result_bahrain_df = pd.read_html(bahrain_race_result_url)[0]
race_result_bahrain_df.columns=['1', 'Position', 'Driver Number', 'Driver', 'Car', 'Laps', 'Time/Retired', 'Points', '2']
race_result_bahrain_df
```
