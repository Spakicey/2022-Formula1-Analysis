```python
# Bahrain Race 2022
# [Pos, No, Driver, Car, Race Laps, Time/Retired, Race PTS], [Fastest Lap, Time of day, Fastest Lap Time, Avg Speed], [Total Stops], [Starting Grid Position, Final Qualifying Time], [Q1 Time, Q2 Time, Q3 Time, Qualifying Laps]
bahrain_df = pd.DataFrame()

# Import race result data
df_0_0 = pd.read_csv("./data/0_0_df.csv")

bahrain_df['Final Race Position'] = df_0_0['Pos']
bahrain_df['No'] = df_0_0['No']
bahrain_df['Driver Name'] = df_0_0['Driver']
bahrain_df['Car'] = df_0_0['Car']
bahrain_df['Race Laps'] = df_0_0['Laps']
bahrain_df['Race Time'] = df_0_0['Time/Retired']
bahrain_df['Race Points'] = df_0_0['PTS']
bahrain_df.set_index('No', inplace=True)

df_0_1 = pd.read_csv("./data/0_1_df.csv")
df_0_1.set_index('No', inplace=True)

bahrain_df = bahrain_df.join(df_0_1[['Lap', 'Time of day', 'Time', 'Avg Speed']], on='No')
bahrain_df.rename(columns={"Lap": "Fastest Lap No", "Time": "Fastest Lap Time"}, inplace=True)

df_0_2 = pd.read_csv("./data/0_2_df.csv")
df_0_2.set_index('No', inplace=True)
df_0_2.sort_values(by=['Driver', 'Stops'])
df_0_2.drop_duplicates(subset=['Driver'], keep='last', inplace=True)

bahrain_df = bahrain_df.join(df_0_2[['Stops', 'Total']], on='No')
bahrain_df.rename(columns={"Total": "Total Pit Time"}, inplace=True)
bahrain_df.rename(columns={"Time": "Fastest Lap Time"}, inplace=True)

df_0_3 = pd.read_csv("./data/0_3_df.csv")
df_0_3.set_index('No', inplace=True)

bahrain_df = bahrain_df.join(df_0_3[['Pos', 'Time']], on='No')
bahrain_df.rename(columns={"Lap": "Fastest Lap No", "Total": "Total Pit Time", "Pos": "Starting Grid Pos", "Time": "Starting Grid Quali Time"}, inplace=True)


df_0_4 = pd.read_csv("./data/0_4_df.csv")
df_0_4.set_index('No', inplace=True)

bahrain_df = bahrain_df.join(df_0_4[['Pos', 'Q1', 'Q2', 'Q3', 'Laps']], on='No')
bahrain_df.rename(columns={"Pos": "Qualifying Postition", "Laps": "Qualifying Laps"}, inplace=True)


df_0_5 = pd.read_csv("./data/0_5_df.csv")
df_0_5.set_index('No', inplace=True)

bahrain_df = bahrain_df.join(df_0_5[['Pos', 'Time', 'Gap', 'Laps']], on='No')
bahrain_df.rename(columns={"Pos": "P3 Postition", "Time": "P3 Time", "Gap": "P3 Gap", "Laps": "P3 Laps"}, inplace=True)


df_0_6 = pd.read_csv("./data/0_6_df.csv")
df_0_6.set_index('No', inplace=True)

bahrain_df = bahrain_df.join(df_0_6[['Pos', 'Time', 'Gap', 'Laps']], on='No')
bahrain_df.rename(columns={"Pos": "P2 Postition", "Time": "P2 Time", "Gap": "P2 Gap", "Laps": "P2 Laps"}, inplace=True)

df_0_7 = pd.read_csv("./data/0_7_df.csv")
df_0_7.set_index('No', inplace=True)

bahrain_df = bahrain_df.join(df_0_6[['Pos', 'Time', 'Gap', 'Laps']], on='No')
bahrain_df.rename(columns={"Pos": "P1 Postition", "Time": "P1 Time", "Gap": "P1 Gap", "Laps": "P1 Laps"}, inplace=True)

bahrain_df['Race'] = 'Bahrain'
bahrain_df.head(5)

```
