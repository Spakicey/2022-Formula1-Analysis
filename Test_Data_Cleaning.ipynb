# Removing unwanted characters

```python
# Remove '+' and 's' from Time/Retired column
df['Time/Retired']= df['Time/Retired'].astype('str')
df['Time/Retired']= df['Time/Retired'].map(lambda x: x.lstrip('+-'))
df['Time/Retired']= df['Time/Retired'].map(lambda x: x.rstrip('s'))
```

```python
# Remove '+' and 's' from Practice 1 Gap column
df['Practice 1 Gap']= df['Practice 1 Gap'].astype('str')
df['Practice 1 Gap']= df['Practice 1 Gap'].map(lambda x: x.lstrip('+-'))
df['Practice 1 Gap']= df['Practice 1 Gap'].map(lambda x: x.rstrip('s'))
```
